* Create a function named preprocess_titanic that accepts the train, validate, and test titanic data, and returns the dataframes ready for modeling.

```python
def preprocess_titanic(train, validate, test):
    train[['is_cherbourg', 'is_queenstown', 'is_southampton']] = pd.get_dummies(train.embark_town).astype(int)
    train = train.drop(columns='embark_town')
    train[['is_female', 'is_male']] = pd.get_dummies(train.sex).astype(int)
    train = train.drop(columns='sex')
    
    validate[['is_cherbourg', 'is_queenstown', 'is_southampton']] = pd.get_dummies(validate.embark_town).astype(int)
    validate = validate.drop(columns='embark_town')
    validate[['is_female', 'is_male']] = pd.get_dummies(validate.sex).astype(int)
    validate = validate.drop(columns='sex')
    
    test[['is_cherbourg', 'is_queenstown', 'is_southampton']] = pd.get_dummies(test.embark_town).astype(int)
    test = test.drop(columns='embark_town')
    test[['is_female', 'is_male']] = pd.get_dummies(test.sex).astype(int)
    test = test.drop(columns='sex')
    
    return train, validate, test
```
